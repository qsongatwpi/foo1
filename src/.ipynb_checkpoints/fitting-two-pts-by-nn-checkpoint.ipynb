__Goal__

We will study how good to fit two points by nn. 


More precisely, we want to learn $f\in L^2(B_1)$ for $B_1 = (-1, 1)$ 
by training a batch of $n=2$ points given by $x = \{x_1, x_2\}\in \mathbb R^n$ with nn function
$$N_\theta(x) = \theta_1 x + \theta_2.$$
The loss function is MSE, i.e.
$$l^2(f, g, x) = \sum_{i=1}^n |f(x_i) - g(x_i)|^2.$$

We say $N_\theta$ has a perfect 2-point fit, if  
$$\inf_\theta l^2(f, N_\theta, x) = 0, \forall x\in \mathbb R^2, f\in L^2(B_1).$$
